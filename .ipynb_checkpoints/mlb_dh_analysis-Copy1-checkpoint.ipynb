```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.impute import SimpleImputer
```

```python
# Load the CSV files into Pandas DataFrames

al_allpositions = pd.read_csv('Resources/AL_wpa_allpositions.csv')
al_dh = pd.read_csv('Resources/AL_wpa_dh.csv')
nl_allpositions = pd.read_csv('Resources/NL_wpa_allpositions.csv')
nl_dh = pd.read_csv('Resources/NL_wpa_dh.csv')
```

```python
# Combine AL DataFrames
al_combined = pd.concat([al_allpositions, al_dh])
al_combined['Improved_WPA'] = al_combined['+WPA'] - al_combined['-WPA']
al_team_stats = al_combined.groupby(['Season', 'Team'])['Improved_WPA'].mean().unstack()

al_combined.head()
```

```python
# Combine NL DataFrames
nl_combined = pd.concat([nl_allpositions, nl_dh])
nl_combined['Improved_WPA'] = nl_combined['+WPA'] - nl_combined['-WPA']
nl_team_stats = nl_combined.groupby(['Season', 'Team'])['Improved_WPA'].mean().unstack()

nl_combined.head()
```

```python
# Prepare data for machine learning
al_team_stats_cleaned = al_team_stats.dropna()
al_X = al_team_stats.index.values.reshape(-1, 1)
al_y = al_team_stats.values
```

```python
# Check if shapes are compatible
print("al_X shape:", al_X.shape)
print("al_y shape:", al_y.shape)
```

```python
# Replace NaN values in al_y with the mean of non-missing values
imputer = SimpleImputer(strategy='mean')
al_y_imputed = imputer.fit_transform(al_y)
```

```python
# Split data into training and testing sets
al_X_train, al_X_test, al_y_train, al_y_test = train_test_split(al_X, al_y_imputed, test_size=0.2, random_state=42)
```

```python
# Build and train a linear regression model for AL
al_model = LinearRegression()
al_model.fit(al_X_train, al_y_train)
al_y_pred = al_model.predict(al_X_test)
```

```python
# Calculate metrics for AL model
al_mse = mean_squared_error(al_y_test, al_y_pred)
al_mae = mean_absolute_error(al_y_test, al_y_pred)
al_r2 = r2_score(al_y_test, al_y_pred)

print("AL Linear Regression Model Metrics:")
print(f"Mean Squared Error: {al_mse}")
print(f"Mean Absolute Error: {al_mae}")
print(f"R-squared: {al_r2}")
```

```python
# Visualize improvement in WPA for different teams in AL
plt.figure(figsize=(20, 9))
al_team_stats.plot(kind='bar', stacked=True, ax=plt.subplot(1, 2, 1))
plt.xlabel('Team')
plt.ylabel('Average Improvement in WPA')
plt.title('Improvement in WPA by Team in AL')
plt.legend(title='Team', loc='upper left')
```

```python
# Visualize improvement in WPA for different teams in NL
plt.figure(figsize=(20, 9))
nl_team_stats.plot(kind='bar', stacked=True, ax=plt.subplot(1, 2, 2))
plt.xlabel('Team')
plt.ylabel('Average Improvement in WPA')
plt.title('Improvement in WPA by Team in NL')
plt.legend(title='Team', loc='upper left')

plt.tight_layout()
plt.show()
```

```python
# Calculate summary statistics for improvement by team in AL and NL
al_team_improvement_stats = al_team_stats.describe()
nl_team_improvement_stats = nl_team_stats.describe()
print("AL Team Improvement Stats:")
print(al_team_improvement_stats)
print("\nNL Team Improvement Stats:")
print(nl_team_improvement_stats)
```
